```python
# Create a bar chart
data = pd.read_csv ("../Project 3 Test/Resources/stations_export.csv")
df = pd.DataFrame(data)

# Count the number of stations in each county
county_counts = df['County'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
county_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Stations by County')
plt.xlabel('County')
plt.ylabel('Number of Stations')
plt.xticks(rotation=90)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('stations_by_county.png')

# Show the plot
plt.show()
```
